````python
import pickle
import pandas as pd
import numpy as np
import tqdm
````

````python
pkl_filename = "params_a_matrixs_f.pkl"
params_a_matrixs_f_load = pd.read_pickle(pkl_filename)

params_a_matrixs_f_load
````

````python
gains_cols = ['K θ', 'K θ_dot', 'K ρ', 'K ρ_dot', 'K φ', 'K φ_dot']
angle_cols = ['θ', 'θ_dot', 'ρ', 'ρ_dot', 'φ', 'φ_dot']
````

````python
def apply_flat(row):
    return row.flatten()
````

````python
fine_gains = params_a_matrixs_f_load[params_a_matrixs_f_load['K'].notnull()]
gains_only_ = fine_gains['K'].apply(apply_flat).apply(pd.Series)
gains_only_.columns = gains_cols
sum_gains = gains_only_.abs().sum(axis=1)
gains_mean = sum_gains.abs().mean()
gains_max = sum_gains.abs().max()
fine_gains = pd.concat([fine_gains, gains_only_], axis=1)
fine_gains["weight 0"] = gains_max + 0.1 - sum_gains
fine_gains["weight 1"] = (gains_mean / (1 + sum_gains**2)) + 1
fine_gains["weight 2"] = 0.5 * np.tanh(4-10*sum_gains/gains_mean) + 1
fine_gains["weight 3"] = fine_gains["weight 1"] + fine_gains["weight 2"]

fine_gains
````

````python
import matplotlib.pyplot as plt
plt.plot(np.sort(fine_gains["weight 0"])[::-1])
plt.show()
plt.plot(np.sort(fine_gains["weight 1"])[::-1])
plt.show()
plt.plot(np.sort(fine_gains["weight 2"])[::-1])
plt.show()
plt.plot(np.sort(fine_gains["weight 3"])[::-1])
plt.show()
````

````python
state_cols = ['θ', 'θ_dot', 'ρ', 'ρ_dot', 'φ', 'φ_dot', 'F_h']
````

````python
maximum = 5000

excessive_selection = (fine_gains[gains_cols[0]] > maximum)
for i in gains_cols:
    excessive_selection = excessive_selection | (fine_gains[i] > maximum) | (fine_gains[i] < -maximum)

fine_gains = fine_gains[~excessive_selection]
````

````python
fine_gains_copy = fine_gains.copy()
train_set = fine_gains_copy.sample(frac=0.75, random_state=0)
test_set = fine_gains_copy.drop(train_set.index)
````

````python
gains_only_train = train_set[gains_cols]
weight_train = train_set["weight 2"]
gains_only_test = test_set[gains_cols]
weight_test = test_set["weight 2"]
````

````python
aj_only_train = train_set['A_j'].apply(apply_flat).apply(pd.Series)
aj_only_test = test_set['A_j'].apply(apply_flat).apply(pd.Series)
````

````python
states_only_train = train_set[state_cols]
states_only_test = test_set[state_cols]
````

````python
from sklearn.neural_network import MLPRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression

from sklearn.multioutput import MultiOutputRegressor

from sklearn.metrics import mean_squared_error
````

````python
regressors = [
    ['RFT', MultiOutputRegressor(RandomForestRegressor()), False],
    # ['RFT weighted', MultiOutputRegressor(RandomForestRegressor()), True],
    ['RFTmae', MultiOutputRegressor(RandomForestRegressor(criterion = "mae")), False],
    ['DTR', MultiOutputRegressor(DecisionTreeRegressor()), False],
    ['MLP', MultiOutputRegressor(MLPRegressor()), True],  # actually quite slow and not so good
    ['ADA', MultiOutputRegressor(AdaBoostRegressor()), False],
    ['KNN', MultiOutputRegressor(KNeighborsRegressor()), False],
    ['LR', MultiOutputRegressor(LinearRegression()), False]
]

regressor_selection = pd.DataFrame(data=regressors, columns = ['name', 'regressor', 'weighted'])

regressor_selection
````

````python
for index, regressor in tqdm.tqdm(regressor_selection.iterrows()):
    if regressor['weighted']:
        regressor['regressor'].fit(states_only_train, gains_only_train, weight_train)
    else:
        regressor['regressor'].fit(states_only_train, gains_only_train)
````

````python
predicted_data = []
for regressor in regressor_selection['regressor']:
    predicted_data.append(regressor.predict(states_only_test))

regressor_selection['prediction'] = predicted_data

regressor_selection
````

````python
error_data = []
quadratic_error_data = []
rmse = []
rmse_weighted = []

for prediction in regressor_selection['prediction']:
    error_data.append((prediction - gains_only_test))
    quadratic_error_data.append(((prediction - gains_only_test)**2))
    rmse.append(np.sqrt(mean_squared_error(gains_only_test, prediction)))
    rmse_weighted.append(np.sqrt(mean_squared_error(gains_only_test, prediction, sample_weight=weight_test)))


regressor_selection['error data'] = error_data
regressor_selection['quadratic error data'] = quadratic_error_data
regressor_selection['rmse'] = rmse
regressor_selection['rmse weighted'] = rmse_weighted


regressor_selection
````

````python
mean = []
std = []

for error_data in regressor_selection['error data']:
    mean.append(error_data.mean())
    std.append(error_data.std())

quad_mean = []
quad_std = []

for error_data in regressor_selection['quadratic error data']:
    quad_mean.append(np.array(error_data).mean())
    quad_std.append(np.array(error_data).std())

regressor_selection['mean error'] = mean
regressor_selection['std error data'] = std
regressor_selection['mean quadratic error data'] = quad_mean
regressor_selection['std quadratic error data'] = quad_std

regressor_selection
````

````python
def print_errors(errors, out):
    out.data += ("#### mean\n```\n" + str(errors.mean()) + "\n```\n")
    out.data += ("#### std\n```\n" + str(errors.std()) + "\n```\n")
````

````python
from IPython.display import Markdown as md

out = md("# Error Comparison\n")

for index, row in regressor_selection.head().iterrows():
    out.data += ("## " + row['name'] + "\n")
    out.data += ('### error data\n')
    print_errors(row['error data'], out)
    out.data += ('### quadratic error data' + "\n")
    print_errors(np.array(row['quadratic error data']), out)

out
````
